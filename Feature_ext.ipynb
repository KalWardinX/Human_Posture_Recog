```python
# data -> joints_train.json, joints_test.json
# not done
import os
import cv2
import mediapipe as mp
import json

model_path = './pose_landmarker.task'
BaseOptions = mp.tasks.BaseOptions
PoseLandmarker = mp.tasks.vision.PoseLandmarker
PoseLandmarkerOptions = mp.tasks.vision.PoseLandmarkerOptions
VisionRunningMode = mp.tasks.vision.RunningMode

options = PoseLandmarkerOptions(
    base_options=BaseOptions(model_asset_path=model_path),
    running_mode=VisionRunningMode.IMAGE)


class_dict = {
    'downdog':0,
    'Bitilasana':1,
    'tree':2,
    'Salamba_Sarvangasana':3,
    'Vasisthasana':4,
    'warrior':5
}
    
def evaluate_joints(detection_result , class_name):
    joint = {}

    mp_pose = mp.solutions.pose
    joint['Head'] = [0,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.NOSE.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.NOSE.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.NOSE.value].z]
    ]
    joint['Neck'] = [1,
        [(detection_result.pose_landmarks[0][mp_pose.PoseLandmark.NOSE.value].x + detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_SHOULDER.value].x + detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x)/3,
        (detection_result.pose_landmarks[0][mp_pose.PoseLandmark.NOSE.value].y + detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_SHOULDER.value].y + detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y)/3,
        (detection_result.pose_landmarks[0][mp_pose.PoseLandmark.NOSE.value].z + detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_SHOULDER.value].z + detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_SHOULDER.value].z)/3]
    ]
    joint['ShoulderLeft'] = [2,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_SHOULDER.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_SHOULDER.value].z]
    ]
    joint['ShoulderRight'] = [3,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_SHOULDER.value].z]
    ]
    joint['ElbowLeft'] = [4,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_ELBOW.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_ELBOW.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_ELBOW.value].z]
    ]
    joint['ElbowRight'] = [5,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_ELBOW.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_ELBOW.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_ELBOW.value].z]
    ]
    joint['WristLeft'] = [6,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_WRIST.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_WRIST.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_WRIST.value].z]
    ]
    joint['WristRight'] = [7,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_WRIST.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_WRIST.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_WRIST.value].z]
    ]
    joint['HandLeft'] = [8,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_INDEX.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_INDEX.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_INDEX.value].z]
    ]
    joint['HandRight'] = [9,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_INDEX.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_INDEX.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_INDEX.value].z]
    ]
    joint['HipLeft'] = [10,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_HIP.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_HIP.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_HIP.value].z]
    ]
    joint['HipRight'] = [11,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_HIP.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_HIP.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_HIP.value].z]
    ]
    joint['KneeLeft'] = [12,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_KNEE.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_KNEE.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_KNEE.value].z]
    ]
    joint['KneeRight'] = [13,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_KNEE.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_KNEE.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_KNEE.value].z]
    ]
    joint['AnkleLeft'] = [14,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_ANKLE.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_ANKLE.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_ANKLE.value].z]
    ]
    joint['AnkleRight'] = [15,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_ANKLE.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_ANKLE.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_ANKLE.value].z]
    ]
    joint['FootLeft'] = [16,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.LEFT_FOOT_INDEX.value].z]
    ]
    joint['FootRight'] = [17,
        [detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].x,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].y,
        detection_result.pose_landmarks[0][mp_pose.PoseLandmark.RIGHT_FOOT_INDEX.value].z]
    ]
    joint['output'] = class_dict[class_name]
    return joint



f_train = open(f"./joints/joints_train.json",'w')
jsontrain_data = []

for classes in os.listdir('data/train'):
    for image in os.listdir(f'./data/train/{classes}'):
        # read image
        img = cv2.imread(f'./data/train/{classes}/{image}')
        # Load the input image from an image file.
        try:
          mp_image = mp.Image.create_from_file(f'./data/train/{classes}/{image}')
        except:
          continue
        # resize image
        # Perform pose landmarking on the provided single image.
        # The pose landmarker must be created with the image mode.
        with PoseLandmarker.create_from_options(options) as landmarker:
            pose_landmarker_result = landmarker.detect(mp_image)
            if pose_landmarker_result.pose_landmarks == []:
                continue
            joint = evaluate_joints(pose_landmarker_result, classes)
            jsontrain_data.append(joint)
json.dump(jsontrain_data, f_train)
f_train.close()

f_test = open(f"./joints/joints_test.json",'w')
jsontest_data = []

for classes in os.listdir('data/test'):
    for image in os.listdir(f'./data/test/{classes}'):
        # read image
        img = cv2.imread(f'./data/test/{classes}/{image}')
        # Load the input image from an image file.
        try:
            mp_image = mp.Image.create_from_file(f'./data/test/{classes}/{image}')
        except:
            continue

        # resize image
        # Perform pose landmarking on the provided single image.
        # The pose landmarker must be created with the image mode.
        with PoseLandmarker.create_from_options(options) as landmarker:
            pose_landmarker_result = landmarker.detect(mp_image)
            if pose_landmarker_result.pose_landmarks == []:
                continue
            joint = evaluate_joints(pose_landmarker_result, classes)
            jsontest_data.append(joint)
json.dump(jsontest_data, f_test)
f_test.close()

print('done')
```

```python
# joints_train.json/joints_test.json -> featureset_train.txt/ featureset_test.txt

import json
import math


# joints = {
#     'Head': (, y0, z0),
#     'Neck': (x1, y1, z1),
#     'ShoulderLeft': (x2, y2, z2),
#     'ElbowLeft': (x4, y4, z4),
#     'WristLeft': (x6, y6, z6),
#     'HandLeft': (x8, y8, z8),
#     'ShoulderRight': (x3, y3, z3),
#     'ElbowRight': (x5, y5, z5),
#     'WristRight': (x7, y7, z7),
#     'HandRight': (x9, y9, z9),
#     'HipLeft': (x10, y10, z10),
#     'KneeLeft': (x12, y12, z12),
#     'AnkleLeft': (x14, y14, z14),
#     'FootLeft': (x16, y16, z16),
#     'HipRight': (x11, y11, z11),
#     'KneeRight': (x13, y13, z13),
#     'AnkleRight': (x15, y15, z15),
#     'FootRight': (x17, y17, z17)
# }

# segments = [
#     ('Head','Neck')
#     ('Neck','ShoulderLeft')
#     ('Neck','ShoulderRight')
#     ('ShoulderLeft', 'ShoulderRight'),
#     ('ShoulderLeft', 'ElbowLeft'),
#     ('ElbowLeft', 'WristLeft'),
#     ('WristLeft', 'HandLeft'),
#     ('ShoulderRight', 'ElbowRight'),
#     ('ElbowRight', 'WristRight'),
#     ('WristRight', 'HandRight'),
#     ('ShoulderLeft','WristLeft'),
#     ('ShoulderRight','WritstRight')
#     ('ShoulderLeft', 'HipLeft'),
#     ('ShoulderRight', 'HipRight'),
#     ('HipLeft', 'KneeLeft'),
#     ('HipRight', 'KneeRight'),
#     ('KneeLeft', 'AnkleLeft'),
#     ('KneeRight', 'AnkleRight'),
#     ('AnkleLeft', 'FootLeft'),
#     ('AnkleRight', 'FootRight'),
#     ('HipLeft','HipRight'),
#     ('HipLeft','AnkleLeft'),
#     ('HipRight','AnkleRight'),
#     ('FootLeft','FootRight')

# ]
def get_joint_features(data):
    joint_features= []
    joint_features += data['Head'][1]
    joint_features += data['Neck'][1]
    joint_features += data['ShoulderLeft'][1]
    joint_features += data['ShoulderRight'][1]
    joint_features += data['ElbowLeft'][1]
    joint_features += data['ElbowRight'][1]
    joint_features += data['WristLeft'][1]
    joint_features += data['WristRight'][1]
    joint_features += data['HandLeft'][1]
    joint_features += data['HandRight'][1]
    joint_features += data['HipLeft'][1]
    joint_features += data['HipRight'][1]
    joint_features += data['KneeLeft'][1]
    joint_features += data['KneeRight'][1]
    joint_features += data['AnkleLeft'][1]
    joint_features += data['AnkleRight'][1]
    joint_features += data['FootLeft'][1]
    joint_features += data['FootRight'][1]
    return joint_features

def calc_segments(data):
    segments = []
    segments = [
        (data['Head'], data['Neck']),
        (data['Neck'], data['ShoulderLeft']),
        (data['Neck'], data['ShoulderRight']),
        (data['ShoulderLeft'], data['ShoulderRight']),
        (data['ShoulderLeft'], data['ElbowLeft']),
        (data['ElbowLeft'], data['WristLeft']),
        (data['WristLeft'], data['HandLeft']),
        (data['ShoulderRight'], data['ElbowRight']),
        (data['ElbowRight'], data['WristRight']),
        (data['WristRight'], data['HandRight']),
        (data['ShoulderLeft'], data['WristLeft']),
        (data['ShoulderRight'], data['WristRight']),
        (data['ShoulderLeft'], data['HipLeft']),
        (data['ShoulderRight'], data['HipRight']),
        (data['HipLeft'], data['KneeLeft']),
        (data['HipRight'], data['KneeRight']),
        (data['KneeLeft'], data['AnkleLeft']),
        (data['KneeRight'], data['AnkleRight']),
        (data['AnkleLeft'], data['FootLeft']),
        (data['AnkleRight'], data['FootRight']),
        (data['HipLeft'],data['HipRight']),
        (data['HipLeft'],data['AnkleLeft']),
        (data['HipRight'],data['AnkleRight']),
        (data['FootLeft'],data['FootRight'])
    ]
    return segments

def calc_distance(p,q):
    dist = math.sqrt((p[0]-q[0])**2 + (p[1]-q[1])**2 + (p[2]-q[2])**2)
    return dist

def calc_dist_features(segments,data):
    distance_features = []
    distance_vectors = []
    distance_vectors = segments + [
        (data["Head"],data["HandLeft"]),
        (data["HandLeft"],data["HipRight"]),
        (data["Head"],data["HandRight"]),
        (data["HandRight"],data["HipLeft"]),
        (data["HipLeft"],data["FootRight"]),
        (data["HipRight"],data["FootLeft"]),
        (data["Head"],data["HipLeft"]),
        (data["Head"],data["HipRight"])
    ]

    for dis_vec in distance_vectors:
        distance_features.append(calc_distance(dis_vec[0][1], dis_vec[1][1]))
    return distance_features

def calc_angle(segment1,segment2):

    segment1_v = [segment1[0][1][0] - segment1[1][1][0], segment1[0][1][1] - segment1[1][1][1], segment1[0][1][2] - segment1[1][1][2]]
    segment2_v = [segment2[0][1][0] - segment2[1][1][0], segment2[0][1][1] - segment2[1][1][1], segment2[0][1][2] - segment2[1][1][2]]
    dot_product = segment1_v[0] * segment2_v[0] + segment1_v[1] * segment2_v[1] + segment1_v[2] * segment2_v[2]
    magnitude_product = math.sqrt(segment1_v[0]**2+segment1_v[1]**2+segment1_v[2]**2)*math.sqrt(segment2_v[0]**2+segment2_v[1]**2+segment2_v[2]**2)
    angle = math.acos(dot_product / magnitude_product)
    return angle

def calc_angle_features(segments):
    angle_features = []

    for i in range(len(segments)):
        for j in range(len(segments)):
            if i<j:
                angle_features.append(calc_angle(segments[i], segments[j]))
    return angle_features

joints_train = json.load(open('./joints/joints_train.json'))
featureset_train = open('./features/featureset_train.txt','w')
for i in joints_train:
    segments = calc_segments(i)
    joint_features = get_joint_features(i)
    distance_features = calc_dist_features(segments,i)
    angle_features = calc_angle_features(segments)
    featureset = joint_features + distance_features + angle_features + [i['output']]
    featureset_train.write(str(featureset)[1:-1])
    featureset_train.write('\n')
featureset_train.close()


joints_test = json.load(open('./joints/joints_test.json'))
featureset_test = open('./features/featureset_test.txt','w')
for i in joints_test:
    segments = calc_segments(i)
    joint_features = get_joint_features(i)
    distance_features = calc_dist_features(segments,i)
    angle_features = calc_angle_features(segments)
    featureset = joint_features + distance_features + angle_features + [i['output']]
    featureset_test.write(str(featureset)[1:-1])
    featureset_test.write('\n')
featureset_test.close()

print('done')
```
